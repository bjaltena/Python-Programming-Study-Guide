## LOOPS

### For Loops
Loops through each item in a container assigning the current list item to a new variable

Example: 
```
studentNames = ["Amanda", "Danielle", "Joseph"]

for student in studentNames:
    print(student + " is in my class!")
```

<br>
<br>

### While Loops
The loop continues while the condition is true, the loop exits once the condition becomes false

Example: 
```
counter = 1

while counter <= 10:
    print ("I love you " + str(counter) + " time(s)!")
    counter = counter + 1

print("Pysch!")
```

````python
counter = 0

# range(4) returns [0, 1, 2, 3]
for steps in range(4):
    counter += steps
    print(steps)

print("\n<----------->\n")

# range(1, 10, 2) returns [1, 3, 5, 7, 9]
for steps in range(1,10,2):
    print(steps)

print("\n<----------->\n")

for steps in [1,2,3,4,5]:
    print(steps)

print("\n<----------->\n")

range = [1,2,3,4,5]
for steps in range:
    print(steps)
````

````python
import turtle

for steps in ["red", "blue", "green", "black"]:
    turtle.color (steps)
    turtle.forward (100)
    turtle.right (90)

turtle.done()
````

````python
# THESE TWO LOOPS DO THE SAME THING IN DIFFERENT WAYS

# a list of student names
studentNames = ["Amanda", "Danielle", "Joseph"]

# WHILE LOOP 
# a counter to go through the list
index = 0

# keep going while the counter is less than the length of the studentNames list (in this case, the length is 3)
while index < len(studentNames):

    # Display a message using the current index in the studentNames list
    print(studentNames[index].capitalize() + " is in my class!")

    # Increase the counter
    index = index + 1


# Create a line between the outputs
print("\n<--------------------------------------------->\n")


# FOR LOOP
# Loop through each name in the list and assign the current value to student
for student in studentNames:

    # Display the message using the student variable
    print(student.capitalize() + " is in my class!")
````

````python
# Initialize a boolean variable 
check = True

# Continue looping until a valid input is given
while check:

    # Ask the user for a number
    counter = int(input("Choose a number between 1 and 10."))

    # Check if the number was outside the range of 1 and 10
    if counter < 1 or counter > 10:

        # Ask them to retry entering the number
        print("You entered " + str(counter) + ". Please choose a number between 1 and 10.\n")
    else:

        # Since their number was valid, make the while loop condition False to exit
        check = False


# Loop until the counter reaches 10
while counter <= 10:

    # Display a nice message
    print ("I love you " + str(counter) + " time(s)!")

    # Increase the counter to ensure the loop is not infinite
    counter = counter + 1

# Display a funny message
print("Pysch!")
````

````python
# Combine multiple concepts together

# create an empty list 
cities = []

while True:
    # Ask the user for a city to add to the list until they are done
    response = input("Enter the name of a city you would like to visit or 'done' to exit the loop.").lower()

    # Check if the response was done
    if response != "done":

        # Add the city name to the list
        cities.append(response.capitalize())
    else:

        # Since they entered done, break out of the loop
        break


# Loop through each of the cities mentioned
for city in cities:

    # Display a message for each city
    print("I would love to visit " + city + "!")
````

````python
# Ask the user for a list of party-goer names
response = input("Enter the names of the party-goers (Example: Alice, Bob, Chad").lower().replace(" ", "")

# Create a list by splitting the string by each comma
names = response.split(",")

# Sort the list in alphabetical order
names.sort()

# Capitalize each item in the list
names = [name.capitalize() for name in names]

# Loop through each of the names mentioned
for name in names:

    # Display a message for each party-goer
    print(name + " had a great time at the party!")

# Display the full list
print("\nFull List: " + ", ".join(names))
````
