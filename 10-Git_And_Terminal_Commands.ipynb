## Git Commands

- **git add**
    - Moves changes from the working directory to the staging area.
    - Examples:
        - git add .
            - adds all files in the current directory
        - git add specificFile1 specificFile2
            - only adds the files mentioned

- **git branch**
    - This command is your general-purpose branch administration tool.
    - Example:
        - git branch
            - Lists all of the files in the local repo

- **git checkout**
    - In addition to checking out old commits and old file revisions, git checkout is also the means to navigate existing branches.
    - Examples:
        - git checkout branchName
            - switches the local files to the specified branchName
        - git checkout -b newBranch currentBranch
            - creates a new branch based on the files in the current branch

- **git clone**
    - Creates a copy of an existing Git repository. Cloning is the most common way for developers to obtain a working copy of a remote repository.
    - Example:
        - git clone remoteRepo
            - copies the files of a remote repo into your local system and sets up git

- **git commit**
    - Takes the staged snapshot and commits it to the project history.
    - Example: 
        - git commit -m "a good comment"
            - finalizes all of the staged files

- **git fetch**
    - Fetching downloads a branch from another repository, along with all of its associated commits and files. But, it doesn't try to integrate anything into your local repository. This gives you a chance to inspect changes before merging them with your project.
    - Example: 
        - git fetch
            - Retrieves files for other branches

- **git merge**
    - A powerful way to integrate changes from divergent branches. After forking the project history with git branch, git merge lets you put it back together again.
    - Example:
        - git merge branchName
            - merges the files from branchName into your current branch

- **git pull**
    - Pulling is the automated version of git fetch. It downloads a branch from a remote repository, then immediately merges it into the current branch.
    - Example: 
        - git pull
            - merges code from the remote repo into your current branch

- **git push**
    - Push lets you move a local branch to another repository, which serves as a convenient way to publish contributions.
    - Example:
        - git push
            - moves all of your changes to the remote repo

- **git reset**
    - Undoes changes to files in the working directory. Resetting lets you clean up or completely remove changes that have not been pushed to a public repository.
    - Examples:
        - git reset .
            - Undoes all of the work in the current folder
        - git reset specificFile1 specificFile2
            - Undoes all updates for those 2 files

- **git status**
    - Displays the state of the working directory and the staged snapshot. You’ll want to run this in conjunction with git add and git commit to see exactly what’s being included in the next snapshot.
    - Example:
        - git status
            - see what updates have been made in the working tree and what is currently staged

### Standard Git Flow

```
# Checkout the branch you want to use for your files and updates
git checkout desiredBranch

# Merge in any updates from the remote repo
git pull

# Complete any and all work

# Check which files have been updated
git status

# Add all files or specific files if you do not want to store all of the files remotely
git add .

# Commit the files to your local repo
git commit -m "a comment that reflects your updates"

# Share your code with the remote repo
git push
```
